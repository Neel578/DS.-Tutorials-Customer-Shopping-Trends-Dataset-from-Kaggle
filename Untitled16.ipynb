```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Python\Data Science\Practical-11\shopping_trends.csv")  
```

```python
print(df.head())
```

```python
plt.figure(figsize=(8,5))
avg_purchase_age = df.groupby("Age")["Purchase Amount (USD)"].mean().sort_values()
avg_purchase_age.plot(kind="barh", color="skyblue")
plt.title("Average Purchase Amount by Age")
plt.xlabel("Average Purchase Amount (USD)")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(6,6))
freq_counts = df["Frequency of Purchases"].value_counts()
plt.pie(freq_counts, labels=freq_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Shopping Frequency Distribution")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8,5))
avg_purchase_season = df.groupby("Season")["Purchase Amount (USD)"].mean()
avg_purchase_season.plot(kind="line", marker="o", color="green")
plt.title("Average Purchase Amount by Season")
plt.xlabel("Season")
plt.ylabel("Average Purchase Amount (USD)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df["Purchase Amount (USD)"], bins=20, kde=True, color="orange")
plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Purchase Amount (USD)", hue="Gender", data=df, alpha=0.7)
plt.title("Purchase Amount vs Age")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```
